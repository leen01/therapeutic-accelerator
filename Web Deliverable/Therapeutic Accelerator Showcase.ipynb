```python
import json
dataframe['externalids'] = dataframe['externalids'].apply(json.loads)

def remove_none_fields_from_json_column(df, column_name):
    def remove_none_fields(json_data):
        return {key: value for key, value in json_data.items() if value is not None}
    
    df[column_name] = df[column_name].apply(lambda x: remove_none_fields(x) if pd.notnull(x) else x)
    return df

def format_json_as_string(df, column_name):
    def format_json(json_data):
        if pd.notnull(json_data):
            formatted_string = ', '.join([f"{key}: {value}" for key, value in json_data.items() if pd.notnull(value)])
            return formatted_string
        return ''
    
    df[column_name] = df[column_name].apply(format_json)
    return df


df = remove_none_fields_from_json_column(dataframe, 'externalids')
df = format_json_as_string(df, 'externalids')
```
