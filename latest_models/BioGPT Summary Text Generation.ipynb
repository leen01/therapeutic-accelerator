```python
### Build out BioGpt with Tensorflow
!pip install sacremoses
biogpttokenizer = BioGptTokenizer.from_pretrained("microsoft/biogpt")
biogptmodel = BioGptForCausalLM.from_pretrained("microsoft/biogpt")
```

```python
sample_abstract = abstract_df['abstract'][0]
tokens = biogpttokenizer(sample_abstract, return_tensors="pt")
output = biogptmodel.generate(tokens['input_ids'],
                        num_beams=2,
                        no_repeat_ngram_size=2,
                        top_p=75,
                        top_k=3,
                        min_length=100,
                        max_length=400)
print(biogpttokenizer.decode(output[0], skip_special_tokens=True, clean_up_tokenization_spaces=True))
```
