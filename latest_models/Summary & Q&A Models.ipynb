```python
from transformers import T5ForQuestionAnswering
qa_model = T5ForQuestionAnswering.from_pretrained("t5-large")
```

```python
question_context = sq_dataframe["abstract"].iloc[0]
print(question_context)
print(sample_question)
encoded_text = T5token([sample_question + question_context], return_tensors="tf")
# proceed = np.all(encoded_text["input_ids"][0] > 0)
# proceed
generated_response = qa_model.generate(
    encoded_text.input_ids,
    num_beams=3,
    no_repeat_ngram_size=2,
    top_k=10,
    max_length=50,
    min_length=25,
)
print(
    T5token.decode(
        generated_response[0],
        skip_special_tokens=True,
        clean_up_tokenization_spaces=True,
    )
)
```
