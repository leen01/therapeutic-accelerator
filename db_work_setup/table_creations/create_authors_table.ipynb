```python
# Query to pull in corpus IDs
sql = '''
    SELECT authors FROM attributes;
'''

# Context manager to open the connection to the database and execute the query. 
with engine.connect() as conn:
    query = conn.execute(text(sql))
    
# Get all records and store in dataframe
df = pd.DataFrame(query.fetchall())
df.head()
```

```python
def clean_authors(author): 
    author = ''.join(author.strip('"').split('\\'))
    author = author.strip('"')
    author = author.strip("{}")
    author = ''.join(["[", author.strip('"').replace('","', ","), "]"])
    # author = json.loads(author)
    return author
```

```python
# partition dataframe to make it easier and faster to work with
# doesn't overload the memory this way

ddf = dd.from_pandas(df, npartitions=1000)
authors = ddf.authors.apply(clean_authors, meta=('authors', 'string')).compute()
authors = pd.DataFrame(authors)
new_list = []
for i,v in authors.iterrows():
    try:
        new_list += json.loads(v.authors)
    except: 
        continue
    
authors = pd.DataFrame(new_list)
print(authors.shape)
authors = authors.drop_duplicates()
print(authors.shape)
```

```python
# Upload authors data table to sql
with engine.connect() as conn: 
    authors.to_sql('authors', con = conn, index = False, if_exists = 'replace',chunksize= 10000)
```

```python
# Query to pull in corpus IDs
sql = '''
    SELECT * FROM authors;
'''

# Context manager to open the connection to the database and execute the query. 
with engine.connect() as conn:
    query = conn.execute(text(sql))
    
# Get all records and store in dataframe
df = pd.DataFrame(query.fetchall())
df.head()
```
